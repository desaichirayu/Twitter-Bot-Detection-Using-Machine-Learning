```python
uid = suser_ds2['id'].unique()
tweets = stweet_ds1_full['text']
def is_nan(x):
    return (x is np.nan or x != x)
# to get all tweets made by a particular user
tweets_scam = []
for i in uid:
    twt_indx= stweet_ds1_full[stweet_ds1_full['user_id'] == i].index
    tweet_str = ""
    for j in twt_indx:
        if(not is_nan(tweets[j])):
            tweet_str = tweet_str + " " + tweets[j]
    if(not is_nan(tweet_str)):
       
        x = tweet_str.split(' ')
        str = ""
        #removing urls and user references from tweets
        for i in x:
            url1 = "https:"
            url2 = "http:"
            mentions= "@"
            if (url1 in i) or (url2 in i) or (mentions in i) or (len(i) == i.isdigit()):
                continue
            else:
                valids = ""
                # to retain just letter in the string     
                for character in i:
                    if character.isalpha():
                        valids += character
                str= str+ " " + valids
        
        tweets_scam.append(str)
```

```python
dataset['target'] = y_ds
# cleaning data to remove nan
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)

clean_ds = clean_dataset(dataset)
```

```python
y_ds = dataset['target']
clean_ds = clean_ds.drop(columns=['target'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report,accuracy_score, f1_score

x_train, x_test, y_train, y_test = train_test_split(dataset, y_ds, test_size=0.33)
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)
predictions = logisticRegr.predict(x_test)
score = logisticRegr.score(x_test, y_test)
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
print('Recall:', recall_score(y_test, predictions,average="micro")*100,"%")
print('Precision:', precision_score(y_test, predictions,average="micro")*100,"%")
```
